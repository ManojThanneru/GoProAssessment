# Data Analyst Jobs

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

### Extracting data from csv and store it in a data frame 

```python
jobs_df= pd.read_csv("DataAnalyst.csv", index_col = 0)
jobs_df.head()
```

### Removing columns which are not necessary

```python
jobs_df=jobs_df.drop(['Job Description'],axis=1)
```

## EDA

### Change location column to only state short names

```python
jobs_df['state'] = jobs_df['Location'].apply(lambda x : x.split(',')[1])
jobs_df.head()
```

### cleaning the head quarter city name and head quarter state name

```python
jobs_df['Headquarter_city'] = jobs_df['Headquarters'].apply(lambda x : x.split(',')[0])
jobs_df.head()
```

```python
jobs_df['Headquarter_state'] = jobs_df['Headquarters'].apply(lambda x : x.split(',')[-1])
jobs_df.head()
```

```python
jobs_df['state'].unique()
```

```python
jobs_df['Headquarter_state'].unique()
```

```python
jobs_df['Headquarter_city'].unique()
```

```python
jobs_df['state'] = jobs_df['state'].replace('Arapahoe', 'CO', regex=True)
```

### Removing the rows which have salary estimate of -1

```python
jobs_df = jobs_df[jobs_df['Salary Estimate'] != '-1']
```

### calculating the maximum and minimum salary 

```python
salary = jobs_df['Salary Estimate'].apply(lambda x : x.split('(')[0])

# Removing K and dollar sign from salary
rm_kd = salary.apply(lambda x : x.replace('K','').replace('$',''))
rm_kd
```

```python
# Creating columns for minimum, maximum, and average salary
jobs_df['min_salary'] = rm_kd.apply(lambda x : int(x.split('-')[0]))
jobs_df['max_salary'] = rm_kd.apply(lambda x : int(x.split('-')[1]))
jobs_df['avg_salary'] = (jobs_df.min_salary + jobs_df.max_salary) / 2
jobs_df.head()
```

```python
jobs_df=jobs_df.drop(['Salary Estimate'],axis=1)
```

```python
jobs_df.head()
```

### Replacing -1 in competitors and Easy Apply columns with NA

```python
jobs_df['Competitors']=jobs_df['Competitors'].apply(lambda x : x.replace('-1','NA'))
jobs_df['Easy Apply']=jobs_df['Easy Apply'].apply(lambda x : x.replace('-1','NA'))
```

```python
jobs_df.head()
```

### Dropping columns which are not used for further analysis

```python
jobs_df=jobs_df.drop(['Location'],axis=1)
```

```python
jobs_df=jobs_df.drop(['Founded','Revenue'],axis=1)
```

```python
jobs_df.to_csv('jobs.csv',index=False)
```
